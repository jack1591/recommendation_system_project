### –†–∞–±–æ—Ç–∞ —Å –¥–∞–Ω–Ω—ã–º–∏

```python
feed_data = feed_data[feed_data['action'] == 'view']
```

```python
import nltk
nltk.download('wordnet')
```

```python
import re
import string

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

def preprocessing(line, token = wnl):
    line = line.lower()
    line = re.sub(r"[{}]".format(string.punctuation), " ", line)   
    line = line.replace('\n\n', ' ').replace('\n', ' ')
    line = ' '.join([token.lemmatize(x) for x in line.split(' ')])
    return line

tfidf = TfidfVectorizer(
    stop_words = 'english',
    preprocessor = preprocessing
)
```

```python
tfidf_data = (
    tfidf
    .fit_transform(post_data['text'])
    .toarray()
)
tfidf_data
```

```python
tfidf_data = pd.DataFrame(
    tfidf_data,
    index = post_data['post_id'],
    columns = tfidf.get_feature_names_out()
)
tfidf_data
```

```python
# –≥–µ–Ω–µ—Ä–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ tf-idf
post_data['SumTfIdf'] = tfidf_data.sum(axis = 1).reset_index()[0]
post_data['MaxTfIdf'] = tfidf_data.max(axis = 1).reset_index()[0]
post_data['MeanTfIdf'] = tfidf_data.mean(axis = 1).reset_index()[0]
post_data
```

```python
[f"DistanceTo{i}thCluster" for i in range(1,16)]
```

```python
# –∫–ª–∞—Å—Ç–µ—Ä–∏–∑—É–µ–º —Ç–µ–∫—Å—Ç—ã
from sklearn.decomposition import PCA

centered = tfidf_data - tfidf_data.mean()

pca = PCA(n_components = 20)
pca_decomp = pca.fit_transform(centered)

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 15, random_state = 0)
kmeans.fit(pca_decomp)

post_data['TextCluster'] = kmeans.labels_

dist_columns = ['DistanceTo1thCluster',
 'DistanceTo2thCluster',
 'DistanceTo3thCluster',
 'DistanceTo4thCluster',
 'DistanceTo5thCluster',
 'DistanceTo6thCluster',
 'DistanceTo7thCluster',
 'DistanceTo8thCluster',
 'DistanceTo9thCluster',
 'DistanceTo10thCluster',
 'DistanceTo11thCluster',
 'DistanceTo12thCluster',
 'DistanceTo13thCluster',
 'DistanceTo14thCluster',
 'DistanceTo15thCluster']

dists_df = pd.DataFrame(
    data = kmeans.transform(pca_decomp),
    columns = dist_columns
)
dists_df
```

```python
post_data
```

```python
post_data = pd.concat((post_data, dists_df), axis = 1)
post_data
```

```python
df = pd.merge(feed_data,
             post_data,
             on = 'post_id',
             how = 'left')
df
```

```python
df = pd.merge(df,
             user_data,
             on = 'user_id',
             how = 'left')
df
```

```python
df['hour'] = pd.to_datetime(df['timestamp']).dt.hour
df['day'] = pd.to_datetime(df['timestamp']).dt.day
df['month'] = pd.to_datetime(df['timestamp']).dt.month
df
```

```python
df = df.drop(['action', 'text'], axis = 1)
df = df.set_index(['user_id', 'post_id'])
df
```

### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

```python
# –¥–∞–Ω–Ω—ã–µ –∏–º–µ—é—Ç –≤—Ä–µ–º–µ–Ω–Ω—É—é —Å—Ç—Ä—É–∫—Ç—É—Ä—É, –∑–Ω–∞—á–∏—Ç –º–æ–∂–Ω–æ –ø—Ä–æ–≤–∞–ª–∏–¥–∏—Ä–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏

df_train = df[df['timestamp'] < '2021-12-15']
df_test = df[df['timestamp'] >= '2021-12-15']

X_train = df_train.drop(['timestamp','target'], axis = 1)
y_train = df_train['target']

X_test = df_test.drop(['timestamp','target'], axis = 1)
y_test = df_test['target']

y_train.shape, y_test.shape
```

```python
X_train
```

```python
# –Ω–∞—á–Ω–µ–º —Å —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞
from sklearn.compose import ColumnTransformer
from category_encoders.one_hot import OneHotEncoder
from category_encoders import TargetEncoder

object_cols = ['topic', 'TextCluster', 'gender', 'country', 'city', 'exp_group', 'os', 'source', 'hour', 'day', 'month']
cols_for_ohe = [col for col in object_cols if X_train[col].nunique()<5]
cols_for_mte = [col for col in object_cols if X_train[col].nunique()>=5]

cols_for_ohe_idx = [list(X_train.columns).index(col) for col in cols_for_ohe]
cols_for_mte_idx = [list(X_train.columns).index(col) for col in cols_for_mte]

t = [
    ('OneHotEncoder', OneHotEncoder(), cols_for_ohe_idx),
    ('MeanTargetEncoder', TargetEncoder(), cols_for_mte_idx)
]

col_transform = ColumnTransformer(transformers = t)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline

pipe = Pipeline([('column_transformer', col_transform),
                ('decision_tree', DecisionTreeClassifier())])

pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

print(f'–ö–∞—á–µ—Å—Ç–≤–æ –Ω–∞ —Ç—Ä–µ–π–Ω–µ: {round(roc_auc_score(y_train, pipe.predict_proba(X_train)[:, 1]),3)}')
print(f'–ö–∞—á–µ—Å—Ç–≤–æ –Ω–∞ —Ç–µ—Å—Ç–µ: {round(roc_auc_score(y_test, pipe.predict_proba(X_test)[:, 1]),3)}')
```

```python
# –≤—Ç–æ—Ä–∞—è –º–æ–¥–µ–ª—å - –±—É—Å—Ç–∏–Ω–≥
from catboost import CatBoostClassifier

catboost = CatBoostClassifier(
    iterations = 500,
    max_depth = 2,
    learning_rate = 1
)
catboost.fit(X_train, y_train, object_cols)
```

```python
print(f'–ö–∞—á–µ—Å—Ç–≤–æ –Ω–∞ —Ç—Ä–µ–π–Ω–µ: {round(roc_auc_score(y_train, catboost.predict_proba(X_train)[:, 1]),3)}')
print(f'–ö–∞—á–µ—Å—Ç–≤–æ –Ω–∞ —Ç–µ—Å—Ç–µ: {round(roc_auc_score(y_test, catboost.predict_proba(X_test)[:, 1]),3)}')
```

```python
def plot_feature_importances(importance, names, model_type):
    feature_importance = np.array(importance)
    feature_names = np.array(names)

    data = {"feature_names": feature_names,
            "feature_importance": feature_importance}

    data = pd.DataFrame(data)

    data.sort_values(by = ['feature_importance'], ascending = False, inplace = True)

    figure = plt.figure()
    figure.set_size_inches(16,10)

    sns.barplot(x = data['feature_importance'], y = data['feature_names'])
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
    plt.title(f"{model_type}_FEATURE_IMPORTANCE")
    plt.show()

plot_feature_importances(catboost.feature_importances_, X_train.columns, 'catboost')
```

```python
# —Å–æ—Ö—Ä–∞–Ω–∏–º –º–æ–¥–µ–ª—å
catboost.save_model(
    'catboost_model',
    format = 'cbm'
)
```
